```python
from scipy.stats import bernoulli
import seaborn as sb

data_bern = bernoulli.rvs(size=10,p=0.125)
ax = sb.distplot(data_bern,
                  kde=True,
                  color='cyan',
                  hist_kws={"linewidth": 25,'alpha':1})
ax.set(xlabel='Bernouli', ylabel='Frequency')
```

```python
from scipy.stats import poisson
import seaborn as sns
import matplotlib.pyplot as plt

#Generating Poisson data
sample1 = poisson.rvs(mu=5, size=100)
sample2 = poisson.rvs(mu=2, size=100)
# sample3 = poisson.rvs(mu=10, size=100)

#Plotting Poisson Distribution using Seaborn
sns.kdeplot(x=sample1, fill=True, label='lambda=5')
sns.kdeplot(x=sample2, fill=True, label='lambda=2')
# sns.kdeplot(x=sample3, fill=True, label='lambda=10')
plt.xlabel('Num of events in a specific time interval')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

The probability density function (pdf) for an exponential distribution is given by the equation:

$f(x, \lambda) = \lambda.e^{-\lambda.x}$   $x=>0$

where:

$\lambda$ = rate at which an event occurs

$X$ = random variable (time between 2 events)

$f(x; \lambda)$ = probability of time between 2 events being x units



Lets plot an exponential distribution using Python:

Q. Plot exponential distributions given that the average time between two successive messages is 50, 60 and 70 seconds.

```python
from scipy.stats import expon
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

lambda1 = 40
lambda2 = 70
min_success = 30

m = (lambda1 + lambda2)/2
x = np.linspace(0, 2*m, 1000)
x_fail = np.linspace(0, min_success, 1000)

# x1_fail = np.linspace(0, lambda1, 1000)
# x2_fail = np.linspace(0, lambda2, 1000)
# print(x1_fail)

#When average time between 2 collision is 50 seconds
y1 = expon.rvs(scale=lambda1, size=10000)

#When average time between 2 collision is 60 seconds
y2 = expon.rvs(scale=lambda2, size=10000)

print(y1)
# y1_fail = expon.rvs(scale=m, size=10000)
# y2_fail = expon.rvs(scale=m, size=10000)
# plt.fill_between(y1_fail,y2_fail,color='b')

#Plot sample data
sns.kdeplot(x=y1, fill=False, label='1/$\lambda_1$=70', color ='g')
ax = sns.kdeplot(x=y2, fill=False, label='1/$\lambda_2$=40', color ='b')
ax.axvline(x = min_success,     
           linestyle='dashed',
             color='r',
               alpha=0.5, 
               ymin=0.025, ymax=0.995, 
               label=r"$R_{min} = $" + f"{min_success}") 
# sns.kdeplot(x=data3, fill=True, label='1/lambda=80')
# plt.xlabel('Units of time between successive events')
plt.xlabel('Rewards:  ' + r'$R_1, \; R_2$')
plt.ylabel('Probability')
plt.title('Exponential Distribution')
plt.legend()
plt.xlim(0, 200)
plt.axvline(lambda1, color='b', linewidth=1)
plt.axvline(lambda2, color='g', linewidth=1)
# plt.axvline(x=min_success, linestyle='dashed', color='r', alpha=0.5, ymin=0.025, ymax=0.995, label=r"$R_{min} = $" + f"{min_success}", zorder=30)
plt.title(r"$P\left(R_2 < R_{min}\right) \; \ll  \; P\left(R_1 < R_{min}\right)$")
plt.savefig('Exp_distribution.eps', format='eps', dpi = 300)
plt.show();
```
